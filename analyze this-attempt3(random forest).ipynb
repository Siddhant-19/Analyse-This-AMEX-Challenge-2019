```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression 
```

```python
data=pd.read_csv("development_dataset.csv")
```

```python
data.head()
```

```python
var=pd.read_excel("data_dictionary.xlsx")
var
```

```python
var.dropna(inplace=True)
```

```python
variables=pd.DataFrame(var["Variable Name"])
variables.head()
```

```python
data.columns=[var["Variable Name"]]
data.replace(to_replace =".", 
                 value =0,inplace=True)
data['total_cards'] = data.total_cards.astype(float)
data.head(5)
```

```python
data.replace(to_replace ="Low", 
                 value =1.0,inplace=True) 
data.replace(to_replace ="Medium", 
                 value =2.0,inplace=True) 
data.replace(to_replace ="High", 
                 value =3.0,inplace=True)
```

```python
data.head()
data.info()
```

```python
X=data.drop(["Assigned Line"],axis=1)
X.head()
```

```python
Y=data["Assigned Line"]
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
X_train.head()
```

```python
X_train.describe()
mean_X_train=X_train.mean()
mean_X_train
```

```python
X_train.fillna(mean_X_train,inplace=True)
X_test.fillna(mean_X_train,inplace=True)
y_train.head()
```

```python
from sklearn.ensemble import RandomForestRegressor 
  
 # create regressor object 
randomreg = RandomForestRegressor(n_estimators = 100, random_state = 0) 
  
# fit the regressor with x and y data 
randomreg.fit(X_train, y_train['Assigned Line'].to_numpy().ravel())
```

```python
y_pred = randomreg.predict(X_test)
```

```python
y_pred=pd.DataFrame(y_pred)
y_pred=y_pred.round()
y_pred.head()
```

```python
from sklearn.metrics import accuracy_score 
#print ("Accuracy : ", accuracy_score(y_test, y_pred))
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
cm = confusion_matrix(y_test, y_pred) 
  
print ("Confusion Matrix : \n", cm)
print(classification_report(y_test,y_pred))
```
